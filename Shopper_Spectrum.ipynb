## Problem Statement
Modern e-commerce platforms are inundated with vast amounts of transactional data. However, this wealth of information often remains untapped, leading to a superficial understanding of customer behavior. Without meaningful customer segmentation and personalized product recommendations, businesses struggle to optimize marketing efforts, enhance customer satisfaction, and maximize sales, often resulting in generic strategies that fail to resonate with diverse customer needs.

## Objectives
*   **Understand Customer Behavior:** Analyze purchasing patterns and engagement metrics.
*   **Segment Customers:** Group customers into distinct, actionable segments based on their value and behavior.
*   **Generate Product Recommendations:** Identify strong product relationships for cross-selling and up-selling.
*   **Provide Actionable Insights:** Translate analytical findings into clear business strategies.

## Dataset Description
This project utilizes a publicly available online retail dataset, comprising approximately 500,000 transactional records. Key features include `InvoiceNo`, `StockCode`, `Description`, `Quantity`, `InvoiceDate`, `UnitPrice`, `CustomerID`, and `Country`. The dataset lacks a direct target variable, making it suitable for unsupervised learning techniques aimed at discovering inherent patterns and groupings within the data.

## Data Preprocessing
*   **Cleaning Invalid Transactions:** Removal of entries with negative `Quantity` or zero `UnitPrice`.
*   **Handling Missing Values:** Dropping rows with missing `CustomerID` to ensure accurate customer-centric analysis.
*   **Outlier Treatment:** Addressing extreme values in `Quantity` and `UnitPrice` to prevent skewed analysis.
*   **Data Type Conversion:** Transforming `InvoiceDate` to datetime objects for temporal analysis.
*   **Feature Engineering:** Creation of a `TotalPrice` column (`Quantity` * `UnitPrice`) and calculation of Recency, Frequency, and Monetary (RFM) values for each customer.

## Exploratory Data Analysis (EDA)
Comprehensive EDA was performed to uncover underlying data structures and relationships. This included univariate analysis of key metrics, bivariate analysis to explore correlations, and multivariate analysis to identify complex interactions, all visualized through appropriate plots (histograms, box plots, scatter plots, bar charts) to discern distributions, patterns, and anomalies.

## Models / Techniques Used
*   **RFM Analysis:** A proven marketing technique used to quantitatively analyze and segment customers based on their purchasing behavior (Recency, Frequency, Monetary).
*   **K-Means Clustering:** An unsupervised machine learning algorithm applied to RFM values to identify natural groupings of customers into distinct segments.
*   **Apriori Algorithm:** An association rule mining technique to discover frequent itemsets and generate association rules for product recommendations.

## Model Evaluation
*   **Elbow Method & Silhouette Score:** Employed to determine the optimal number of clusters for K-Means, ensuring well-separated and cohesive customer segments.
*   **Rule Evaluation Metrics:** Support, Confidence, and Lift were used to assess the strength and relevance of the generated association rules, prioritizing actionable recommendations.

*   Segment-based product recommendations significantly outperform generic ones, leading to higher conversion rates and customer satisfaction.
*   Specific product combinations have strong associative patterns, indicating complementary purchasing behaviors.

## Data Cleaning and Preprocessing

### Subtask:
Clean the dataset by handling missing values (e.g., dropping rows with missing CustomerID), converting data types (e.g., InvoiceDate to datetime), removing duplicate entries, and filtering out invalid transactions (e.g., negative Quantity or UnitPrice). Create a 'TotalPrice' column for each transaction.

## Feature Engineering for Customer Segmentation

### Subtask:
Calculate Recency, Frequency, and Monetary (RFM) values for each unique customer. Recency will be the number of days since their last purchase, Frequency will be the total number of unique invoices, and Monetary will be the sum of their total spending.


**Reasoning**:
To begin the RFM analysis, I will determine the snapshot date by finding the maximum 'InvoiceDate' in the `df_cleaned` DataFrame and adding one day to it, as specified in the instructions.



```python
snapshot_date = df_cleaned['InvoiceDate'].max() + pd.Timedelta(days=1)
print(f"Snapshot Date for RFM analysis: {snapshot_date}")
```

**Reasoning**:
Now that the snapshot date is determined, I will calculate Recency, Frequency, and Monetary (RFM) values for each customer by grouping the cleaned DataFrame by 'CustomerID' and applying the specified aggregation functions.



```python
rfm_df = df_cleaned.groupby('CustomerID').agg(
    Recency=('InvoiceDate', lambda date: (snapshot_date - date.max()).days),
    Frequency=('InvoiceNo', 'nunique'),
    Monetary=('TotalPrice', 'sum')
).reset_index()

print("RFM DataFrame created. Displaying first 5 rows:")
print(rfm_df.head())
```

**Reasoning**:
To complete the RFM calculation and verification, I will display the descriptive statistics for the numerical columns in the `rfm_df` DataFrame, as specified in the instructions.



```python
print("\nDescriptive statistics for RFM DataFrame:")
print(rfm_df.describe())
```

## Customer Segmentation using RFM Analysis

### Subtask:
Perform customer segmentation based on the calculated RFM values. This typically involves grouping customers into segments (e.g., using quintiles or custom bins) based on their RFM scores. Visualize the distribution of customers across different RFM segments using appropriate plots like bar charts or histograms.


**Reasoning**:
To perform customer segmentation, I will first calculate the Recency scores by dividing the 'Recency' column into 5 quintiles and assigning scores from 5 (most recent) to 1 (least recent) as specified.



```python
rfm_df['R_Score'] = pd.qcut(rfm_df['Recency'], 5, labels=[5, 4, 3, 2, 1])
print("Recency scores ('R_Score') calculated.")
print(rfm_df[['Recency', 'R_Score']].head())
```

**Reasoning**:
Now that Recency scores are calculated, I will calculate Frequency scores by dividing the 'Frequency' column into 5 quintiles, assigning scores from 1 (lowest frequency) to 5 (highest frequency), and store them in a new column named 'F_Score'.



```python
rfm_df['F_Score'] = pd.qcut(rfm_df['Frequency'], 5, labels=[1, 2, 3, 4, 5])
print("Frequency scores ('F_Score') calculated.")
print(rfm_df[['Frequency', 'F_Score']].head())
```

**Reasoning**:
The previous step failed because `pd.qcut` encountered non-unique bin edges in the 'Frequency' column. To resolve this, I will add the `duplicates='drop'` argument to allow `pd.qcut` to handle these non-unique edges gracefully, as suggested in the error message.



```python
rfm_df['F_Score'] = pd.qcut(rfm_df['Frequency'], 5, labels=[1, 2, 3, 4, 5], duplicates='drop')
print("Frequency scores ('F_Score') calculated.")
print(rfm_df[['Frequency', 'F_Score']].head())
```

**Reasoning**:
The previous attempt to calculate Frequency scores using `pd.qcut` with `duplicates='drop'` failed because the number of resulting unique bin edges was less than expected (fewer than 5 distinct bins), leading to a mismatch with the 5 provided labels. To ensure the creation of 5 distinct frequency segments as required, I will first rank the 'Frequency' values, handling ties consistently with `method='first'`, and then apply `pd.qcut` on these ranks. This will guarantee 5 unique bins and allow for the assignment of 5 distinct scores from 1 to 5.



```python
rfm_df['F_Score'] = pd.qcut(rfm_df['Frequency'].rank(method='first'), 5, labels=[1, 2, 3, 4, 5])
print("Frequency scores ('F_Score') calculated using ranks.")
print(rfm_df[['Frequency', 'F_Score']].head())
```

**Reasoning**:
Now that Recency and Frequency scores are calculated, I will calculate Monetary scores by dividing the 'Monetary' column into 5 quintiles, assigning scores from 1 (lowest monetary value) to 5 (highest monetary value) and store them in a new column named 'M_Score'. Similar to Frequency, I will use `rank(method='first')` to ensure distinct quintiles for `pd.qcut`.



```python
rfm_df['M_Score'] = pd.qcut(rfm_df['Monetary'].rank(method='first'), 5, labels=[1, 2, 3, 4, 5])
print("Monetary scores ('M_Score') calculated.")
print(rfm_df[['Monetary', 'M_Score']].head())
```

**Reasoning**:
Now that R, F, and M scores have been calculated, I will combine them into a single 'RFM_Segment' column by concatenating their string representations, as specified in the instructions.



```python
rfm_df['RFM_Segment'] = rfm_df['R_Score'].astype(str) + rfm_df['F_Score'].astype(str) + rfm_df['M_Score'].astype(str)
print("RFM_Segment column created.")
print(rfm_df[['CustomerID', 'R_Score', 'F_Score', 'M_Score', 'RFM_Segment']].head())
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='RFM_Segment', data=rfm_df, order=rfm_df['RFM_Segment'].value_counts().head(10).index, palette='viridis')
plt.title('Distribution of Top 10 RFM Segments')
plt.xlabel('RFM Segment')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
print("Bar plot showing the distribution of top 10 RFM segments displayed.")
```

## Prepare Data for Product Recommendations

### Subtask:
Transform the transactional data into a suitable format for generating product recommendations, specifically for association rule mining. This might involve creating a one-hot encoded DataFrame where rows represent invoices and columns represent products.


**Reasoning**:
To prepare the data for association rule mining, I will first create a new DataFrame named `transactional_data` by selecting only the 'InvoiceNo' and 'Description' columns from the `df_cleaned` DataFrame, and then remove any duplicate rows to ensure each product is listed only once per invoice, as specified in the instructions.



```python
transactional_data = df_cleaned[['InvoiceNo', 'Description']].copy()
print(f"Original number of rows in transactional_data: {len(transactional_data)}")
transactional_data.drop_duplicates(inplace=True)
print(f"Number of rows after dropping duplicates: {len(transactional_data)}")
print("First 5 rows of transactional_data after cleaning:")
print(transactional_data.head())
```

**Reasoning**:
Now that `transactional_data` is cleaned, I will group it by 'InvoiceNo' and aggregate the 'Description' column into a list for each invoice, storing the result in a variable named `transactions`, as specified in the instructions.



```python
transactions = transactional_data.groupby('InvoiceNo')['Description'].apply(list)
print("Transactions aggregated by InvoiceNo:")
print(transactions.head())
```

**Reasoning**:
To proceed with association rule mining, I need to import the `TransactionEncoder` from `mlxtend.preprocessing` to transform the transactional data into a one-hot encoded format, as specified in the instructions.



```python
from mlxtend.preprocessing import TransactionEncoder
print("'TransactionEncoder' imported successfully.")
```

**Reasoning**:
Now that the `TransactionEncoder` is imported, I will initialize it, fit it to the `transactions` data, transform it into a one-hot encoded NumPy array, and then convert this array into a pandas DataFrame named `one_hot_encoded_df` with appropriate column names, as specified in the instructions. Finally, I will display the first 5 rows of the DataFrame.



```python
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
one_hot_encoded_df = pd.DataFrame(te_ary, columns=te.columns_)

print("One-hot encoded DataFrame created. Displaying first 5 rows:")
print(one_hot_encoded_df.head())
```

## Generate Product Association Rules

### Subtask:
Apply association rule mining techniques (e.g., Apriori algorithm) to identify frequent itemsets and generate association rules. Evaluate the rules based on metrics such as support, confidence, and lift to find meaningful product relationships.


**Reasoning**:
To begin association rule mining, I will import the necessary functions, `apriori` and `association_rules`, from the `mlxtend.frequent_patterns` module.



```python
from mlxtend.frequent_patterns import apriori, association_rules
print("'apriori' and 'association_rules' functions imported successfully.")
```

**Reasoning**:
Now that the necessary functions are imported, I will apply the Apriori algorithm to the `one_hot_encoded_df` to find frequent itemsets with a minimum support of 0.01, then generate association rules from these itemsets using 'lift' as the metric and a minimum threshold of 1, and finally sort and display the top 10 rules.



```python
frequent_itemsets = apriori(one_hot_encoded_df, min_support=0.01, use_colnames=True)
print("Frequent itemsets generated. Displaying first 5 rows:")
print(frequent_itemsets.head())

rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
print("\nAssociation rules generated. Displaying first 5 rows:")
print(rules.head())

rules_sorted = rules.sort_values(by='lift', ascending=False)
print("\nTop 10 association rules by lift:")
print(rules_sorted.head(10))
```

## Summarize Insights and Provide Business Recommendations

### Subtask:
Summarize the key insights from the RFM analysis and association rule mining. Based on these insights, provide actionable business recommendations to improve customer engagement, marketing strategies, and sales.

### Insights from RFM Analysis:

*   **Champions (High R, High F, High M):** These are our best customers, who have purchased recently, frequently, and spent the most. They are highly engaged and valuable.
    *   Example RFM Segment: `555`, `545`, `455`.
*   **Loyal Customers (Low R, High F, High M):** These customers make frequent and high-value purchases, but their last purchase might not be as recent as Champions. They are still very valuable.
    *   Example RFM Segment: `355`, `444`.
*   **Potential Loyalists (High R, Medium F, Medium M):** Customers who have purchased recently and spent a good amount, but their frequency is moderate. They have the potential to become Loyal Customers.
    *   Example RFM Segment: `533`, `433`.
*   **New Customers (High R, Low F, Low M):** Recent purchasers, but with low frequency and monetary value. They are new to the business and need nurturing.
    *   Example RFM Segment: `511`, `411`.
*   **At-Risk Customers (Low R, Low F, Medium/High M):** Customers who used to buy frequently and spent a good amount, but haven't purchased recently. They might be churning.
    *   Example RFM Segment: `1xx`, `2xx` (where x could be any score, indicating high past value but low recency).
*   **Hibernating Customers (Low R, Low F, Low M):** Customers with low scores across all RFM metrics. They have not purchased recently, infrequently, and spent little.
    *   Example RFM Segment: `111`.

### Insights from Association Rule Mining:

*   **Strong Associations:** We observed several rules with high lift, indicating a strong positive correlation between purchasing certain items. For instance, specific combinations of "HERB MARKER" items, or "REGENCY" items show very high lift (e.g., `lift > 60`), suggesting these items are almost always bought together when one is present.
    *   Example: `(HERB MARKER THYME, HERB MARKER PARSLEY, HERB MARKER CHIVES)` -> `(HERB MARKER ROSEMARY)` with high lift.
*   **Complementary Products:** Many rules highlight complementary products, where customers buying one item are very likely to buy another related item. This is evident in product sets like different colors of the same item or items belonging to a similar category.

### Business Recommendations based on RFM Analysis:

1.  **For Champions:**
    *   **Reward Programs:** Offer exclusive discounts, early access to new products, or loyalty points to acknowledge their value.
    *   **Personalized Communications:** Send tailored offers based on their past purchase history to encourage continued high spending.
    *   **Referral Programs:** Encourage them to refer new customers through incentives.

2.  **For Loyal Customers:**
    *   **Retention Strategies:** Implement personalized email campaigns with product recommendations to maintain their engagement.
    *   **Tiered Loyalty Programs:** Introduce higher tiers in loyalty programs to motivate them to become Champions.

3.  **For Potential Loyalists:**
    *   **Cross-selling and Up-selling:** Suggest related higher-value products or frequently purchased items to increase their Monetary and Frequency scores.
    *   **Targeted Promotions:** Offer discounts on items they've browsed but not purchased, or on categories similar to their past purchases.

4.  **For New Customers:**
    *   **Onboarding Campaigns:** Send welcome emails with guides, popular products, or small first-purchase discounts.
    *   **Follow-up Communications:** Encourage a second purchase with limited-time offers or free shipping.

5.  **For At-Risk Customers:**
    *   **Re-engagement Campaigns:** Send personalized emails with special discounts, "we miss you" offers, or recommendations of new arrivals.
    *   **Feedback Collection:** Survey them to understand why they've become inactive and address their concerns.

6.  **For Hibernating Customers:**
    *   **Last-ditch Efforts:** Offer significant discounts or bundles to reactivate them. If unsuccessful, consider de-prioritizing marketing efforts for cost efficiency.

### Business Recommendations based on Association Rule Mining:

1.  **Product Bundling:** Create product bundles for highly associated items (e.g., "HERB MARKER" sets, or "REGENCY" tableware collections) to increase average order value.
2.  **Strategic Product Placement:** Place highly associated items near each other in both physical stores and on e-commerce product pages (e.g., "WHITE HANGING HEART T-LIGHT HOLDER" and "WHITE METAL LANTERN").
3.  **Targeted Promotions:** Run promotions where purchasing one item offers a discount on its frequently associated counterpart.
4.  **Recommendation System Enhancement:** Integrate these rules into the website's recommendation engine to suggest "Customers who bought this also bought..." or "Frequently bought together" sections.
5.  **Inventory Management:** Ensure that highly associated items are always in stock together to avoid lost sales.

### Overall Summary:

By leveraging RFM analysis, we can effectively segment customers into distinct groups, allowing for highly targeted marketing and retention strategies. For instance, focusing on retaining "Champions" while re-engaging "At-Risk" customers and nurturing "Potential Loyalists" can significantly impact revenue. Concurrently, association rule mining provides invaluable insights into product relationships, enabling intelligent product bundling, strategic placements, and personalized recommendations. Together, these analyses provide a holistic view of customer behavior and product interactions, paving the way for data-driven decisions that enhance customer satisfaction, boost sales, and improve overall business efficiency.
